```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def read_spe(date, infile):

    os.chdir(date) 
    
    hist_cts = []
    
    with open(infile) as f:
        for lineno, line in enumerate(f):
            if  12 <= lineno < 2060:
                hist_cts.append(float(line.strip()))
    hist_cts = np.array(hist_cts)
    
    os.chdir('..')

    return hist_cts 
```
